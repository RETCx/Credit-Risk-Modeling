# Expected loss (EL) and its components: PD, LGD and EAD 

Expected Loss (EL) เป็นแนวคิดสำคัญในการจัดการความเสี่ยงด้านเครดิต และหมายถึงจำนวนเงินเฉลี่ยที่ผู้ให้กู้คาดว่าจะสูญเสียหากผู้กู้ไม่สามารถชำระหนี้ได้ โดยมีองค์ประกอบหลัก 3 ส่วนคือ:

Probability of Default (PD): คือความน่าจะเป็นที่ผู้กู้จะผิดนัดชำระหนี้ในช่วงระยะเวลาหนึ่ง โดยปกติจะคำนวณในระยะเวลา 1 ปี ค่า PD แสดงเป็นเปอร์เซ็นต์ ซึ่งได้รับอิทธิพลจากปัจจัยต่าง ๆ เช่น ประวัติเครดิตของผู้กู้ สภาพเศรษฐกิจ และปัจจัยเสี่ยงอื่น ๆ

Loss Given Default (LGD): แสดงถึงเปอร์เซ็นต์ของยอดหนี้ทั้งหมดที่สูญเสียไปเมื่อผู้กู้ผิดนัดชำระหนี้ หลังจากหักมูลค่าที่อาจได้รับคืนจากหลักประกันหรือการชดใช้ค่าเสียหายอื่น ๆ แล้ว ค่า LGD จะแสดงเป็นเปอร์เซ็นต์และมักจะน้อยกว่า 100% เพราะผู้ให้กู้อาจได้รับเงินบางส่วนคืนจากการขายหลักประกันหรือวิธีอื่น ๆ

Exposure at Default (EAD): คือจำนวนเงินทั้งหมดที่เสี่ยงต่อการสูญเสียเมื่อเกิดการผิดนัดชำระหนี้ ค่า EAD รวมถึงยอดหนี้คงค้างของเงินกู้รวมถึงดอกเบี้ยหรือค่าธรรมเนียมอื่น ๆ ที่ผู้กู้ค้างชำระในเวลาที่ผิดนัด

การคำนวณ Expected Loss (EL) ใช้สูตรดังนี้:
EL=PD×LGD×EAD
สูตรนี้ให้ค่าการสูญเสียทางการเงินที่คาดว่าจะเกิดจากการผิดนัดชำระหนี้ โดยพิจารณาจากความน่าจะเป็นของการผิดนัด (PD) จำนวนเงินที่เสี่ยงต่อการสูญเสีย (EAD) และความรุนแรงของการสูญเสียเมื่อเกิดการผิดนัด (LGD)

# Basel II

Basel II ถูกใช้เป็นแนวทางในการกำกับดูแลความเสี่ยงในธนาคารทั่วโลก และเป็นพื้นฐานในการพัฒนามาตรฐาน Basel III ที่มีการปรับปรุงเพิ่มเติมเพื่อรับมือกับปัญหาทางการเงินที่เกิดขึ้นในภายหลัง

## Capital adequacy, regulations, and the Basel II accord

Capital Adequacy หมายถึง ความสามารถของธนาคารในการรักษาระดับเงินทุนให้เพียงพอเพื่อรองรับความสูญเสียที่อาจเกิดขึ้นและเพื่อให้สามารถดำเนินกิจการต่อไปได้ในช่วงที่เกิดความผันผวนทางการเงิน

Regulations เกี่ยวกับความเพียงพอของเงินทุนมีความสำคัญอย่างมาก เพราะช่วยให้มั่นใจว่าธนาคารมีฐานะทางการเงินที่แข็งแกร่งพอที่จะรับมือกับความเสี่ยงและป้องกันไม่ให้เกิดปัญหาทางการเงินที่อาจส่งผลกระทบต่อระบบการเงินโดยรวม ซึ่งหน่วยงานกำกับดูแลมักจะกำหนดข้อกำหนดขั้นต่ำเกี่ยวกับเงินทุนที่ธนาคารต้องมีเพื่อให้ดำเนินการได้อย่างมั่นคง

Basel II Accord หรือ ข้อตกลงบาเซิล II เป็นมาตรฐานสากลที่กำหนดโดยคณะกรรมการ Basel Committee on Banking Supervision (BCBS) ซึ่งมุ่งเน้นการปรับปรุงความแข็งแกร่งทางการเงินของธนาคารโดยการกำหนดข้อกำหนดขั้นต่ำสำหรับเงินทุนและความเสี่ยง Basel II ถูกพัฒนาขึ้นมาเพื่อเสริมสร้างความมั่นคงในระบบการเงินโลกและประกอบด้วย 3 เสาหลักคือ:

Minimum Capital Requirements (ข้อกำหนดขั้นต่ำของเงินทุน): กำหนดให้ธนาคารต้องมีเงินทุนขั้นต่ำเพียงพอที่จะรองรับความเสี่ยงที่เกิดจากสินทรัพย์ต่าง ๆ ที่ถือครองอยู่

Supervisory Review Process (กระบวนการตรวจสอบของหน่วยงานกำกับดูแล): เน้นการตรวจสอบและประเมินความเสี่ยงของธนาคารอย่างเป็นระบบ โดยให้หน่วยงานกำกับดูแลสามารถตรวจสอบและกำหนดมาตรการเพิ่มเติมตามความเหมาะสม

Market Discipline (วินัยตลาด): ส่งเสริมความโปร่งใสในการเปิดเผยข้อมูลทางการเงินของธนาคาร เพื่อให้ผู้มีส่วนได้ส่วนเสียสามารถประเมินความเสี่ยงและสภาพคล่องของธนาคารได้อย่างถูกต้อง

## Basel II approaches: SA, F-IRB, and A-IRB


ภายใต้ข้อตกลง Basel II ธนาคารสามารถเลือกใช้วิธีการประเมินความเสี่ยงด้านเครดิตที่แตกต่างกันเพื่อคำนวณความต้องการเงินทุนขั้นต่ำ โดยแบ่งเป็น 3 วิธีหลัก ได้แก่ Standardized Approach (SA), Foundation Internal Ratings-Based Approach (F-IRB), และ Advanced Internal Ratings-Based Approach (A-IRB):





### Standardized Approach (SA)



ในวิธีนี้ ธนาคารจะใช้การจัดอันดับเครดิตจากหน่วยงานจัดอันดับภายนอก (External Credit Rating Agencies) เพื่อกำหนดความเสี่ยงของสินทรัพย์แต่ละรายการ เช่น สินเชื่อให้แก่ลูกค้าหรือเงินลงทุนในตราสารหนี้
ความเสี่ยงที่กำหนดจะถูกแปลงเป็นน้ำหนักความเสี่ยง (Risk Weight) ซึ่งใช้คำนวณเงินทุนที่ธนาคารต้องถือไว้
วิธีนี้เหมาะสำหรับธนาคารที่ไม่ต้องการหรือไม่สามารถพัฒนาระบบประเมินความเสี่ยงภายในได้

### Internal Ratings-Based Approach (IRB)

Foundation Internal Ratings-Based Approach (F-IRB)

ธนาคารที่ใช้วิธีนี้จะสามารถพัฒนาระบบการจัดอันดับความเสี่ยงภายใน (Internal Ratings) เพื่อตรวจวัดความน่าจะเป็นของการผิดนัดชำระหนี้ (Probability of Default, PD) ของลูกค้าหรือสินทรัพย์
อย่างไรก็ตาม ธนาคารจะต้องใช้ค่าที่กำหนดโดย Basel II สำหรับ Loss Given Default (LGD) และ Exposure at Default (EAD) แทนที่จะคำนวณเอง
วิธีนี้เปิดโอกาสให้ธนาคารที่มีความสามารถในการจัดทำระบบประเมินความเสี่ยงภายในระดับหนึ่งสามารถใช้ประโยชน์ จากข้อมูลของตัวเองเพื่อคำนวณความเสี่ยงได้

Advanced Internal Ratings-Based Approach (A-IRB)

ในวิธีนี้ ธนาคารจะพัฒนาระบบการประเมินความเสี่ยงที่ครอบคลุมมากขึ้น โดยสามารถคำนวณและใช้ค่าความเสี่ยงภายในเองสำหรับ PD, LGD, และ EAD
วิธีนี้มอบความยืดหยุ่นสูงสุดให้กับธนาคารในการใช้ข้อมูลภายในและแบบจำลองเฉพาะทางเพื่อประเมินความเสี่ยงและคำนวณความต้องการเงินทุนขั้นต่ำ
วิธีนี้เหมาะสำหรับธนาคารขนาดใหญ่ที่มีทรัพยากรและความเชี่ยวชาญในการจัดทำแบบจำลองความเสี่ยงภายในที่ซับซ้อน
