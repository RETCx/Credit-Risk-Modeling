### meaning of loan_status

Fully Paid: การชำระเงินของเงินกู้ได้ถูกจ่ายเต็มจำนวนโดยผู้กู้แล้ว และไม่มีการชำระเงินเพิ่มเติมที่ค้างอยู่

Charged Off: ผู้ให้กู้ได้ตัดหนี้ออกเป็นขาดทุนเพราะเชื่อว่าผู้กู้จะไม่ชำระหนี้คืน ซึ่งมักจะเกิดขึ้นหลังจากช่วงเวลานานของการไม่ชำระเงิน

Current: เงินกู้ได้รับการชำระตามกำหนดเรียบร้อยแล้ว และผู้กู้กำลังปฏิบัติตามข้อผูกพันการชำระเงินทั้งหมดตามกำหนด

Default: ผู้กู้ล้มเหลวในการชำระเงินตามข้อตกลงเงินกู้ และเงินกู้ถือว่ามีความผิดพลาดร้ายแรง การนิยามของการผิดนัดอาจแตกต่างกันไป แต่โดยทั่วไปหมายถึงการละเมิดข้อตกลงเงินกู้ที่สำคัญ

Late (31-120 days): การชำระเงินของเงินกู้ล่าช้า 31 ถึง 120 วัน สถานะนี้หมายความว่าผู้กู้ได้พลาดการชำระเงิน แต่ยังไม่ถึงระดับการผิดนัด

In Grace Period: เงินกู้อยู่ในช่วงระยะเวลาผ่อนผันที่ผู้กู้พลาดการชำระเงิน แต่ได้รับเวลามากขึ้นในการชำระเงินโดยไม่ต้องเผชิญกับบทลงโทษหรือผิดนัด

Late (16-30 days): การชำระเงินของเงินกู้ล่าช้า 16 ถึง 30 วัน สถานะนี้หมายถึงความผิดพลาดในระยะสั้นที่ผู้กู้ได้พลาดการชำระเงินหนึ่งหรือมากกว่าหนึ่งครั้ง แต่ยังไม่ถึงระดับการผิดนัดที่ร้ายแรง

Does not meet the credit policy. Status: Fully Paid: สถานะนี้หมายความว่าแม้ว่าเงินกู้จะได้รับการชำระเต็มจำนวน แต่ไม่ตรงตามข้อกำหนดนโยบายเครดิตในบางช่วงเวลาของชีวิตเงินกู้ ซึ่งอาจบ่งบอกว่าเงินกู้ได้รับการชำระแล้ว แต่ถูกจัดอยู่ในปัญหานโยบายเครดิตระหว่างช่วงเวลาหนึ่ง

Does not meet the credit policy. Status: Charged Off: สถานะนี้หมายความว่าเงินกู้ถูกตัดออกเป็นขาดทุนและไม่ตรงตามข้อกำหนดนโยบายเครดิต ซึ่งแสดงว่าเงินกู้ถูกเขียนออกและถูกติดป้ายว่าผิดนโยบายเครดิต

n_obs:

คำอธิบาย: จำนวนของการสังเกตสำหรับแต่ละเกรด

คำอธิบาย: สัดส่วนของเงินกู้ที่ "ดี" (เช่น เงินกู้ที่ชำระเต็มจำนวน) สำหรับแต่ละเกรด
การใช้งาน: ใช้ในการคำนวณจำนวนเงินกู้ดีและไม่ดี และคำนวณสัดส่วนของเงินกู้ดีและไม่ดีสำหรับแต่ละเกรด

```python
df1.columns = ['grade','n_obs','prop_good']
df1
```

```python
df1['prop_n_obs'] = df1['n_obs']/df1['n_obs'].sum()
```

```python
df1
```

n_good:

คำอธิบาย: จำนวนจริงของเงินกู้ดีสำหรับแต่ละเกรด
การใช้งาน: คำนวณจาก prop_good คูณด้วย n_obs ใช้ในการคำนวณสัดส่วนของเงินกู้ดีและไม่ดี

n_bad:

คำอธิบาย: จำนวนจริงของเงินกู้ไม่ดีสำหรับแต่ละเกรด
การใช้งาน: คำนวณเป็น (1 - prop_good) คูณด้วย n_obs ใช้ในการคำนวณสัดส่วนและ WoE

```python
df1['n_good'] = df1['prop_good']*df1['n_obs']
df1['n_bad'] = (1-df1['prop_good']) * df1['n_obs']
df1
```

prop_n_good:

คำอธิบาย: สัดส่วนของเงินกู้ดีสำหรับแต่ละเกรดเทียบกับจำนวนรวมของเงินกู้ดีทั้งหมด
การใช้งาน: ใช้ในการคำนวณ WoE เพื่อเข้าใจว่าแต่ละเกรดมีส่วนร่วมในจำนวนเงินกู้ดีทั้งหมดมากแค่ไหน
prop_n_bad:

คำอธิบาย: สัดส่วนของเงินกู้ไม่ดีสำหรับแต่ละเกรดเทียบกับจำนวนรวมของเงินกู้ไม่ดีทั้งหมด
การใช้งาน: ใช้ในการคำนวณ WoE เพื่อเข้าใจว่าแต่ละเกรดมีส่วนร่วมในจำนวนเงินกู้ไม่ดีทั้งหมดมากแค่ไหน

```python
df1['prop_n_good'] = df1['n_good'] / df1['n_good'].sum()
df1['prop_n_bad'] = df1['n_bad'] / df1['n_bad'].sum()
df1
```

สรุปค่าทั้งหมดเพื่อให้ได้หมายเลขเดียวที่สรุปพลังการคาดการณ์ของตัวแปร ค่าที่สูงกว่า IV มักแสดงถึงตัวแปรที่มีความสามารถในการแยกแยะดีขึ้น

IV สรุปความแข็งแกร่งในการคาดการณ์โดยรวมของตัวแปรประเภทนั้น ๆ

Information Value (IV) คือ ค่าที่ใช้ในการประเมินความสามารถในการคาดการณ์ของตัวแปรในการแยกแยะระหว่างผลลัพธ์ที่ดีและไม่ดีในโมเดลการประเมินความเสี่ยงทางเครดิต โดยเฉพาะในกระบวนการสร้างสกอร์การ์ด (scorecard) และการเลือกฟีเจอร์ที่ใช้ในโมเดล

ความหมายของ IV
ค่าของ IV จะบ่งบอกถึงความสามารถในการคาดการณ์ของตัวแปร ดังนี้:

IV < 0.02: ตัวแปรไม่สำคัญ (not useful)

0.02 ≤ IV < 0.1: ตัวแปรมีความสำคัญน้อย (weak predictor)

0.1 ≤ IV < 0.3: ตัวแปรมีความสำคัญปานกลาง (moderate predictor)

0.3 ≤ IV < 0.5: ตัวแปรมีความสำคัญสูง (strong predictor)
